```python
#set up
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

import pandas as pd
import time
from tqdm import tqdm
import os

# -----------------------------
# Setup paths
# -----------------------------
path_wd = "D:/Projektfolder1 (Miethe, Grosenick)/zz_AmelieMisc/"
path_data = os.path.join(path_wd, "thankyou/data/repec/")

os.chdir(path_wd)


# List of months and years to loop over
months = list(range(1, 13))  # 1,2,...,12
years = list(range(2015, 2026))  # 2015,2016,...,2025
```

```python
# -------------------------------------------------------------
# SCRAPE Top 5% AUTHORS BY MONTH/YEAR
# -------------------------------------------------------------


# Set up Selenium WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

try:
    for year in years:
        for month in months:
            print(f"\nScraping year={year}, month={month}")

            # Navigate to the page (update URL if monthly-specific URL exists)
            yymm = f"{year % 100:02d}{month:02d}"  # RePEc format: YYMM
            url = f"https://ideas.repec.org/top/old/{yymm}/top.person.alldetail.html"            
            driver.get(url)
            time.sleep(3)  # allow page to load

            # Locate table
            table = driver.find_element(By.CSS_SELECTOR, "table#left-cols.newbigtop")
            rows = table.find_elements(By.TAG_NAME, "tr")

            # Extract data
            table_data = []
            for row in tqdm(rows, desc="Processing rows"):
                cells = row.find_elements(By.TAG_NAME, "td") or row.find_elements(By.TAG_NAME, "th")
                cell_texts = [cell.text.strip() for cell in cells]
                if cell_texts:
                    table_data.append(cell_texts)

            # Convert to DataFrame
            if table_data:
                headers = table_data[0]
                df = pd.DataFrame(table_data[1:], columns=headers)
                df["year"] = year
                df["month"] = month

                # Save CSV for this month-year
                yymm = f"{year:02d}{month:02d}"
                csv_file = os.path.join(path_data, f"repec_top5_{yymm}.csv")
                df.to_csv(csv_file, index=False, encoding="utf-8")
                print(f"Saved table to {csv_file}")

finally:
    driver.quit()
```

```python
# -------------------------------------------------------------
# SCRAPE Top 6-10% AUTHORS BY MONTH/YEAR
# -------------------------------------------------------------


# Set up months and years
months = list(range(1, 3))      # Jan–Dec (change to 1-13 for full year)
years = list(range(2015, 2016))  # 2015–2025

# Set up Selenium WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

try:
    for year in years:
        for month in months:
            print(f"\nScraping Top 6-10%: year={year}, month={month}")

            # Navigate to page (update URL if monthly-specific exists)
            url = "https://ideas.repec.org/top/old/1502/top.person.alldetail.html"
            driver.get(url)
            time.sleep(3)  # allow page to load

            try:
                # Find all <h3> elements for Top 6% to Top 10%
                top_headers = driver.find_elements(
                    By.XPATH,
                    "//h3[contains(text(), 'Top 6%') or contains(text(), 'Top 7%') or contains(text(), 'Top 8%') or contains(text(), 'Top 9%') or contains(text(), 'Top 10%')]"
                )

                full_text = []

                for header in top_headers:
                    section_title = header.text.strip()
                    section_texts = [section_title]  # start with the header itself

                    # Collect all following siblings until next <h3>
                    try:
                        elem = header.find_element(By.XPATH, "following-sibling::*")
                        while elem.tag_name != "h3":
                            if elem.text.strip():
                                section_texts.append(elem.text.strip())
                            try:
                                elem = elem.find_element(By.XPATH, "following-sibling::*")
                            except:
                                break  # reached end
                    except:
                        pass  # no following siblings

                    full_text.extend(section_texts)

                # Combine into one string
                combined_text = "\n".join(full_text)

                # Save to TXT file per month-year
                yymm = f"{year % 100:02d}{month:02d}"  # e.g., 1501 for Jan 2015
                txt_file = os.path.join(path_data, f"repec_top6_10_{yymm}.txt")
                with open(txt_file, "w", encoding="utf-8") as f:
                    f.write(combined_text)

                print(f"Saved Top 6-10% text to {txt_file}")

            except Exception as e:
                print(f"Could not find Top 6-10% sections for {year}-{month}: {e}")

finally:
    driver.quit()
```
